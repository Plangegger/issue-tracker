```python
from sqlite3 import  connect
```

```python
connection = connect("issue_tracker.db")
```

```python
einkaufswagen = connection.execute("""
CREATE TABLE issue (
    issue_id INTEGER PRIMARY KEY AUTOINCREMENT,
    name TEXT,
    priority TEXT, 
    deadline DATETIME,
    is_done BOOL,
    description TEXT
);                               
""")

connection.commit()
```

```python
einkaufswagen = connection.execute("""
insert into issue (name, description, deadline) values
('Einkaufen gegen', ' Kaufe Eier, Milch und Brot', ' 2026-04-22');                               
""")

connection.commit()
```

```python
einkaufswagen = connection.execute("""
select * from issue; 
""")
einkaufswagen.fetchall()
```

```python
def add_issue(name, description, deadline):
    connection = connect("issue_tracker.db")
    sql = f"insert into issue (name, description, deadline) values ('{name}', '{description}', '{deadline}');"
    einkaufswagen = connection.execute(sql)
    connection.commit()
    einkaufswagen.close()
    connection.close()
```

```python
add_issue("Blumen kaufen", "Kaufe einen Straß rosen für Mama", "2026-05-01")
```

```python
einkaufswagen = connection.execute("""
select * from issue; 
""")
einkaufswagen.fetchall()
```
